```python
#_EDA_1
#_Which country faces maximum and minimum terrorist attacks
eda1=pd.DataFrame(df.groupby(df["country_txt"])["eventid"].count())
eda1.reset_index(inplace=True)
eda1.rename(columns={"country_txt":"Country","eventid":"No_of_attacks"},inplace=True)
top_5=eda1.sort_values(by='No_of_attacks',ascending=False,na_position='first').head(5)
top_5
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(data=top_5,x=top_5["Country"],y=top_5["No_of_attacks"],hue=top_5["Country"])



```

```python
eda1=pd.DataFrame(df.groupby(df["country_txt"])["eventid"].count())
eda1.reset_index(inplace=True)
eda1.rename(columns={"country_txt":"Country","eventid":"No_of_attacks"},inplace=True)
tail_5=eda1.sort_values(by='No_of_attacks',ascending=True,na_position='first').head(5)
tail_5
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(data=tail_5,x=tail_5["Country"],y=tail_5["No_of_attacks"],hue=tail_5["Country"])
```

```python
#_EDA_2
#_In which year maximum number of attacks happened
year_wise_attacks=pd.DataFrame(df.groupby(df["iyear"])["eventid"].count())
year_wise_attacks.reset_index(inplace=True)
year_wise_attacks.rename(columns={"iyear":"Year","eventid":"No_of_attacks"},inplace=True)
Descending=year_wise_attacks.sort_values(by='No_of_attacks',ascending=False,na_position='first').head(5)
ascending=year_wise_attacks.sort_values(by='No_of_attacks',ascending=True,na_position='first').head(5)

```

```python
Descending # maximum attacks by year
```

```python
sns.heatmap(Descending,annot=True,fmt='d')
```

```python
 # minimum attacks by year
sns.heatmap(ascending,annot=True,fmt='d')
```

```python
#_EDA_3
#_In which month most terrorist activities take place
eda3=pd.DataFrame(df.groupby(['imonth'])["eventid"].count())
eda3.reset_index(inplace=True)
eda3.drop(axis=0,index=0,inplace=True)
```

```python
sns.scatterplot(eda3,x=eda3['imonth'],y=eda3['eventid'])
```

```python
#_EDA_4
#_Number of attacks tracked by GTD
eda_4=pd.DataFrame(df.groupby("dbsource")["eventid"].count())
eda_4.reset_index(inplace=True)
linplt=sns.lineplot(eda_4,x=eda_4['dbsource'],y=eda_4['eventid'])
plt.xticks(rotation=90)
plt.show()
```

```python
#eda_5
#Frequency of Attacks
plt.figure(figsize=(9,4))
sns.countplot(x=df["iyear"], hue="success", data = df )
plt.xlabel("iyear")
plt.xticks(rotation=90)
plt.ylabel("Frequency of Attacks")
```
